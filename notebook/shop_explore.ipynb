```python
count = train.groupby(["shop_id","weekday"]).count()["user_id"].reset_index()
d = zip(shops,range(num_class))
d = dict(d)
matrix = np.zeros((7,num_class))
count["label_index"] = count.shop_id.map(lambda x:d[x])
x = count[["weekday","label_index","user_id"]].values
for _x in x:
    matrix[_x[0],_x[1]] = _x[2]
matrix = pd.DataFrame(matrix)
```

```python
hour_split = [1,2,4,6]
for _h in hour_split:
    train["hour_{}".format(_h)] = (train.hour / _h).astype(int)
```

```python
def shop_attractive_matrix(name,name2,train,test)
    max_v = train[name2].max()
    count = train.groupby([name,name2]).count()["user_id"].reset_index()
    d = zip(shops,range(num_class))
    d = dict(d)
    matrix = np.zeros((max_v,num_class))
    count["label_index"] = count.shop_id.map(lambda x:d[x])
    x = count[[name2,"label_index","user_id"]].values
    for _x in x:
        matrix[_x[0],_x[1]] = _x[2]
    matrix[name2] = range(max_v)
    train_matrix = pd.merge(train[[name2]],matrix,how="left",on=name2).drop(name2,axis=1).values
    test_matrix = pd.merge(test[[name2]],matrix,how="left",on=name2).drop(name2,axis=1).values
    return train_matrix, test_matrix
```
